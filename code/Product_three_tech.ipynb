```python
import pandas as pd
from SCPIntModel import *
```

```python
import scanpy as sc
```

```python
multibatch_ann = sc.read_h5ad('../data/task_data/three_tech_ann.h5ad')
```

```python
multibatch_ann
```

```python
processor = AnnDataProcessor(multibatch_ann, batch_key="batch")
```

```python
model_name = "three_tech_cls-0.1_adv-0.3_mse-0.5_warmup-1_lr-0.01_epoch-1000.pt"
```

```python
trainer = Trainer.load(f"../data/model/saved/{model_name}")
```

```python
X_tensor, _ = processor.to_tensors()
```

```python
bio_emb, batch_emb = trainer.encode(X_tensor) 
```

```python
multibatch_ann.obsm["X_SCPInt"] = bio_emb.numpy()
```

```python
multibatch_ann.obsm["batch_em"] = batch_emb.numpy()
```
